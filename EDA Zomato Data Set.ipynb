```python
import pandas as pd

# Assuming final_df is your DataFrame
cuisine_series = final_df['Cuisines']

# Filter out NaN values
cuisine_series = cuisine_series.dropna()

# Combine all cuisines into a single string
all_cuisines = ', '.join(cuisine_series)

# Split the combined string into individual cuisines
individual_cuisines = all_cuisines.split(', ')

# Create a pandas Series and count the occurrences of each cuisine
cuisine_counts = pd.Series(individual_cuisines).value_counts()

# Get the top 10 cuisines
top_10_cuisines = cuisine_counts.head(10)

print("Top 10 Cuisines:")
print(top_10_cuisines)
```
